```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Częstotliwość sygnału
f = 50  # Hz

# Liczba próbek
N = 65536

# Czas trwania próbki (delta czasu)
T = 1.0 / f

# Tworzenie tablicy czasów
t = np.linspace(0.0, N*T, N)

# Generowanie fali sinusoidalnej
y = np.sin(2 * np.pi * f * t)

```

```python
plt.plot(t, y)
plt.title('Fala sinusoidalna o częstotliwości 50 Hz')
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')
plt.grid(True)
plt.show()
```
