```python
%matplotlib ipympl
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
```

<h3>3</h3>

Ustawianie ilości okresów funkcji

```python
textBox = widgets.Text(
    value='1', 
    description='Wartość:',
    placeholder='Wpisz tutaj',
    disabled=False
)
display(textBox)
```

Zwiększenie ilości cykli branych pod uwagę przy licczeniu transformaty Furiera wpływa pozytywnie na znaleznienie waftości częstotliwości

<h3>1</h3>

```python
# Częstotliwość sygnału
f = 50  # Hz

# Liczba próbek
N = 65536

# Czas trwania próbki (delta czasu)
T = 1.0 / f

tmax = np.pi/2/f*float(textBox.value)

#częstotliwość próbkowania
fs = N/tmax


# Tworzenie tablicy czasów
t = np.linspace(0.0, tmax, N)

# Generowanie fali sinusoidalnej
sinSignal = np.sin(2 * np.pi * f * t)

```

```python
plt.figure()
plt.plot(t, sinSignal)
plt.title('Fala sinusoidalna o częstotliwości 50 Hz')
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')
plt.grid(True)
plt.show()
```

<h3>2</h3>

Wyznaczanie dyskretnej transwormacji fouriera

```python
# Obliczanie DFT sygnału

dft = np.fft.fft(sinSignal)
freq = np.fft.fftfreq(len(sinSignal), tmax/N)


# Wykres widma amplitudowego
plt.figure()
plt.plot(freq[:len(freq)//2], abs(dft[:len(freq)//2]))
plt.xlabel('Częstotliwość [Hz]')
plt.ylabel('Amplituda')
plt.title('Widmo amplitudowe sygnału')
plt.grid(True)
plt.show()
```

Niedokładny wynik spowodowany jest wzięciem pod uwagę małej próbki zygnału

<h3>5</h3>

```python
orgSignal = np.fft.ifft(dft)

# Wykres widma amplitudowego
plt.figure()
plt.plot(t, orgSignal.real, label="Oryginalny sygnał")
plt.plot(t, sinSignal, linestyle=':', label="Sygnał odzyskany")
plt.legend()
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')
plt.title('Porównanie fal')
plt.grid(True)
plt.show()
```

<h3>3</h3>

Generowanie nowych sygnałów

```python
sinSignal50 = np.sin(2 * np.pi * 50 * t)
sinSignal60 = np.sin(2 * np.pi * 60 * t)
```

```python
sumSignal = sinSignal50 + sinSignal60

# Wykres widma amplitudowego
plt.figure(figsize=(10, 6))
plt.plot(t, sumSignal)
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')
plt.title('Fala zsumowana z dwóch sinusoidalnych')
plt.grid(True)
plt.show()
```

Transformacja Fouriera

```python
# Obliczanie DFT sygnału
dft = np.fft.fft(sumSignal)
freq = np.fft.fftfreq(len(sumSignal), tmax/N)


# Wykres widma amplitudowego
plt.figure(figsize=(10, 6))
plt.plot(freq[:len(freq)//2], abs(dft[:len(freq)//2]))
plt.xlabel('Częstotliwość [Hz]')
plt.ylabel('Amplituda')
plt.title('Widmo amplitudowe sygnału')
plt.grid(True)
plt.show()
```
